```python
import pandas as pd

# Replace 'Anodic_combined.csv' with the actual file path and name
file_path = 'Anodic_combined.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Define the time intervals
interval = 1.0  # Interval duration in seconds

# Create an empty DataFrame to store the averaged data
averaged_data = pd.DataFrame(columns=['Time3_2', 'Absolute Energy_2', 'Current_2', 'Friction Coefficient_2'])

# Get the minimum and maximum time values from the dataset
min_time = 0
max_time = df['Time3_2'].max()

start_time = min_time
end_time = start_time + interval
count = 0

while start_time <= max_time:
    mask = (df['Time3_2'] >= start_time) & (df['Time3_2'] <= end_time)
    interval_data = df.loc[mask]
    
    if not interval_data.empty:
        avg_values = interval_data[['Time3_2', 'Absolute Energy_2', 'Current_2', 'Friction Coefficient_2']].mean()
        avg_values['Time3_2'] =  end_time  # Calculate the average time
        averaged_data = pd.concat([averaged_data, avg_values.to_frame().T], ignore_index=True) 
    
    start_time = end_time
    end_time += interval

# Save the averaged data to a new CSV file
averaged_data.to_csv("averaged_data.csv", index=False)
```
