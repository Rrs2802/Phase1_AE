```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Read the CSV file into a DataFrame
input_file = "your_input.csv"
output_file = "normalized_output.csv"

data = pd.read_csv(input_file)

# Separate the numeric columns from the string column
numeric_data = data.iloc[:, :4]
string_data = data.iloc[:, 4]

# Initialize the Min-Max Scaler
scaler = MinMaxScaler()

# Fit and transform the numeric data
normalized_numeric_data = scaler.fit_transform(numeric_data)

# Create a DataFrame with the normalized numeric data
normalized_df = pd.DataFrame(normalized_numeric_data, columns=numeric_data.columns)

# Combine the normalized numeric data with the string data
normalized_data = pd.concat([normalized_df, string_data], axis=1)

# Save the normalized data to a new CSV file
normalized_data.to_csv(output_file, index=False)

print("Normalization complete. Data saved to", output_file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file
csv_file = 'all_combined.csv'
df = pd.read_csv(csv_file)

# Extract the column names
independent_variable = df.columns[0]
dependent_variables = df.columns[1:4]

fig_size = (10, 6)

# Create three scatter plots
for dep_var in dependent_variables:
    plt.figure(figsize=fig_size)
    plt.scatter(df[independent_variable], df[dep_var],alpha=0.9,s=30)
    plt.title(f'Scatter Plot of {dep_var} vs {independent_variable}')
    plt.xlabel(independent_variable)
    plt.ylabel(dep_var)

plt.show()
```
